```python
x = np.arange(-3, 3.1, 0.1)
y1 = x**2
y2 = x**3

plt.plot(x, y1)
plt.plot(x, y2)

plt.legend(["Squares", "Cube"], loc = "lower right", bbox_to_anchor=(1.3, 0.5))
plt.text(0, 0, "(0, 0)")
plt.text(1, 1, "(1, 1)")
plt.title("Expontential Functions")
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.grid() # new

plt.show()
```

```python
# Subplots
```

```python
x = np.arange(1, 10, 0.1)
y = np.sin(x)
y1 = np.cos(x)

plt.plot(x, y)
plt.show()

plt.plot(x, y1)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(x, y)

ax[1].plot(x, y1)

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 3))

ax[0].plot(x, y)
ax[1].plot(x, y1)

plt.show()
```

```python
x = np.arange(1, 10, 0.1)
y = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

fig, ax = plt.subplots(3, 1)

ax[0].plot(x, y)
ax[0].plot(x, y2)

ax[1].plot(x, y2)

ax[2].plot(x, y3)

plt.show()
```

```python
x = np.arange(1, 10, 0.1)
y = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

plt.figure()

plt.subplot(3, 1, 1)
plt.plot(x, y)

plt.subplot(3, 1, 2)
plt.plot(x, y2)

plt.subplot(3, 1, 3)
plt.plot(x, y3)

plt.show()
```

```python
plt.figure()

plt.subplot(2, 3, 1)
plt.plot(x, y)

plt.subplot(2, 3, 3)
plt.plot(x, y2)

plt.subplot(2, 3, 5)
plt.plot(x, y3)

plt.show()
```

```python
fig = plt.figure()

x = np.arange(-1, 1, 0.001)
y = np.arange(-1, 1, 0.001)
x, y = np.meshgrid(x, y)

ax = fig.add_subplot(projection="3d")
ax.plot_surface(x,y, x**2 + y**2)

plt.show()
```
