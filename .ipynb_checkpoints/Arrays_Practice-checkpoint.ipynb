```python
from array import array
def add_matrices(arr1, arr2):
    ''' arr1, arr2 both are 2d-arrays
        output: function should return a resultant 2d array or None'''
    #YOUR CODE GOES HERE
    res_arr = []
    for i in range(len(arr1)):
        if(len(arr1)!=len(arr2)):
            return None
        inner_array = array('l',[])
        for j in range(len(arr1[i])):
            if(len(arr1[i])!=len(arr2[i])):
                return None
            inner_array.append(arr1[i][j] + arr2[i][j])
        res_arr.append(inner_array)
   
    return res_arr
```

```python
arr1 = [array('i', [1, 6, 1]), array('i', [5, 9, 6]), array('i', [1, 4, 1])]
#arr2 = [array('i', [3, 5, 7]), array('i', [4, 6, 10]), array('i', [2, 10, 7])]
arr2 = [array('i', [3, 5, 7]), array('i', [4, 6, 10])]
print(add_matrices(arr1,arr2))
```
