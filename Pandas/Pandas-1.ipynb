```python
a = pd.DataFrame({'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]})
a = a[['age','name']]
a
```

```python
mask_kid = a['age']<18
mask_adult = (a['age']>=18) & (a['age']<50)
mask_senior = (a['age']>=50)
a.loc[mask_kid,'age'] = 'kid'
a.loc[mask_adult,'age'] = 'adult'
a.loc[mask_senior,'age'] = 'senior'
```
