```python
def romanToInt(s: str) -> int:
    number = 0
    roman_dict = {
        'I' : 1,
        'V' : 5,
        'X' : 10,
        'L' : 50,
        'C' : 100,
        'D' : 500,
        'M' : 1000
    }


    for i in s[::-1]:
        if roman_dict.get(i) < number and i!=prev_s :
            number = number - roman_dict.get(i)
        else:
            number = number + roman_dict.get(i)
            prev_s = i

    return number
```

```python
def twoSum(nums, target):

    fi,si = 0,0
    found = False

    for i in range(len(nums)):
        fi = i
        for j in range(i+1,len(nums)):
            si = j

            if nums[fi] + nums[si] == target:
                found = True
                break
        if found:
            break
            
    print(fi,si,nums[fi],nums[si])
```

```python
def wordBreak(dict, s):
    # code here
    ans = []
    word = ''
    sentence = ''
    no_remaining_sentences = False
    collected_words=set()
    is_new_sentence = True

    while(no_remaining_sentences==False):
        
        for i,ch in enumerate(s):
            
            word = word + ch
            if (word in dict and (word not in collected_words or is_new_sentence)):
                
                print("COLLECTED WORDS : ",collected_words)
                if(i<len(s)-1):
                    sentence = sentence + word + " "
                    collected_words.add(word)
                    word=''    
                    is_new_sentence = True
                elif(i==len(s)-1):
                    sentence = sentence + word
                    collected_words.add(word)
                    is_new_sentence = False
            elif (word not in dict and i==len(s)-1):
                sentence=''
                break
            

        if(sentence==''):
            no_remaining_sentences=True
            is_new_sentence = False
        else:
            is_new_sentence = False
            print("SENTENCE", sentence)
            ans.append(sentence)
            sentence=''
            word=''
            

    return ans
```

```python
def wordBreak(dict, s):
        # code here
        def solve(arr,st,l,res):
            if st==len(s):
                res.append(l[:])
                
            for word in arr:
                if s[st:st+len(word)]==word:
                    l.append(word)
                    solve(arr,st+len(word),l,res)
                    l.pop()
                
        
        res=[]
        solve(dict,0,[],res)
        res=[" ".join(x) for x in res]
        return res
```
